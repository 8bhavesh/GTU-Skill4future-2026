```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler, LabelEncoder

df = pd.read_csv("co2_emmision.csv")
print("="*70)
print("HEAD SECTION")
print(df.head())

print("\n")

print("="*70)
print("DATASET INFORMATION")
df.info()

print("\n")

print("="*70)
print("DATASET SHAPE")
print(df.shape)

print("\n")

print("="*70)
print("COLUMN NAMES")
print(df.columns)

print("\n")

print("="*70)
print("STATISTICAL SUMMARY")
print(df.describe())

print("\n")

print("="*70)
print("MISSING VALUES")
print(df.isnull().sum())
# Numerical Columns
num_cols = df.select_dtypes(include=np.number).columns
for col in num_cols:
    df[col] = df[col].fillna(df[col].median())
# Categorical Columns
cat_cols = df.select_dtypes(include="object").columns
for col in cat_cols:
    df[col] = df[col].fillna(df[col].mode()[0])

print("\nMissing Values After Handling")
print(df.isnull().sum())

#  Data Visualization
sns.set_style("whitegrid")
# Histogram
for col in num_cols[:3]:
    plt.figure(figsize=(6,4))
    sns.histplot(df[col], kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()

# Boxplot
for col in num_cols[:3]:
    plt.figure(figsize=(6,4))
    sns.boxplot(x=df[col])
    plt.title(f"Boxplot of {col}")
    plt.show()

# Final Dataset
print("="*70)
print("FINAL DATASET")
print(df.head())

```

General Observations

This dataset contains Carbon CO2 emission information.

Numerical columns represent measurable values.

Categorical columns represent names such as country/region.

Higher population generally increases emissions.

GDP may positively influence emissions.

CO2 emissions may vary across countries and year.

Insights




Missing numerical values were replaced using Median because it is less affected by outliers.

Missing categorical values were replaced using Mode because it represents the most frequent category.

Histograms show the distribution of numerical features.

Boxplots help identify outliers.
