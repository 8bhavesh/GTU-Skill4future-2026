```python
import pandas as pd
df = pd.read_csv("football.csv")
# METHOD 1: head() - First 5 rows
print("="*60)
print("METHOD 1: head() - First 5 rows")
print("="*60)
print(df.head())
print("\n")

# METHOD 2: tail() - Last 5 rows
print("="*60)
print("METHOD 2: tail() - Last 5 rows")
print("="*60)
print(df.tail())
print("\n")

# METHOD 3: info() - Dataset Information
print("="*60)
print("METHOD 3: info() - Dataset Information")
print("="*60)
print(df.info())
print("\n")

# METHOD 4: describe() - Statistical Summary
print("="*60)
print("METHOD 4: describe() - Statistical Summary")
print("="*60)
print(df.describe())
print("\n")

# METHOD 5: shape - Rows and Columns
print("="*60)
print("METHOD 5: shape - Number of rows and columns")
print("="*60)
print(f"Shape: {df.shape}")
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")
print("\n")

# METHOD 6: columns - Column Names
print("="*60)
print("METHOD 6: columns - Column names")
print("="*60)
print(df.columns.tolist())
print("\n")

# METHOD 7: isnull() - Check Missing Values
print("="*60)
print("METHOD 7: isnull() - Check missing values")
print("="*60)
print(df.isnull().sum())
print("\n")

# METHOD 8: dropna() - Remove Missing Values
print("="*60)
print("METHOD 8: dropna() - Remove missing values")
print("="*60)
df_cleaned = df.dropna()
print(f"Original shape: {df.shape}")
print(f"After dropna: {df_cleaned.shape}")
print(df_cleaned.head())
print("\n")

# METHOD 9: value_counts() - Count Unique Values
print("="*60)
print("METHOD 9: value_counts() - Count unique values")
print("="*60)
# Using first column
first_column = df.columns[0]
print(f"Value counts for '{first_column}':")
print(df[first_column].value_counts())
print("\n")

# METHOD 10: sort_values() - Sort Data
print("="*60)
print("METHOD 10: sort_values() - Sort data")
print("="*60)
# Sort by first column
print(f"Sorted by '{first_column}':")
print(df.sort_values(by=first_column).head(10))
```
