```python
import pandas as pd
import numpy as np
df=pd.read_csv("solar_energy.csv")
print("="*80)
print("\nHead:")
print(df.head())

print("="*80)
print("\nMissing Values:")
print(df.isnull().sum())

numeric_cols = df.select_dtypes(include=np.number).columns
for col in numeric_cols:
    df[col] = df[col].fillna(df[col].median())

print("="*80)
print("\nMissing Values After Treatment:")
print(df.isnull().sum())
```

GENERALISATIONS:
1. The dataset contains solar energy-related information for global cities.
2. It includes:
- Geographic data City, Country, Latitude, Longitude, Region
- Solar radiation metrics Annual Sunlight Hours, GHI, Daily Peak Sun Hours
- Financial indicators Electricity Price, System Cost, ROI, Payback Period
- Environmental impact CO2 Reduction per Year
- Investment indicators Solar Installations Count
- Solar Viability Score
3. Cities in Africa, Middle East, and parts of Asia show higher sunlight hours.
4. Europe shows lower solar potential compared to Africa and Australia.
5. Higher sunlight → higher annual production → better ROI.

KEY FINDINGS:
1. Strong positive correlation:
- Sunlight Hours & Production
- Production & Savings
- Savings & ROI
- Production & CO2 Reduction
2. Strong negative correlation:
- Savings & Payback Period
3. Solar Viability Score increases in high-sunlight regions.
4. Since installation cost is fixed,
